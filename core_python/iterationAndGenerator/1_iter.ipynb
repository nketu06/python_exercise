```python
# Iterable → gives iterator
# Iterator → gives values

# Iterable = Factory
# Iterator = Worker

# All iterators are iterable, but not all iterables are iterators.



nums = [1, 2, 3]
num=iter(nums)

print(type(iter(nums))) # equivalent nums.__iter__()
print(num.__next__())  # equivalent next(num)


while True:
    try:
        print(next(num))
    except StopIteration:
        break
```

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self   # iterator returns itself

    def __next__(self):

        if self.current < self.limit:
            self.current += 1
            return self.current

        raise StopIteration

c = Counter(3)

for x in c:
    print(x)

#Iterator is stateful.
c = Counter(3) # need to re-initiate
print([i for i in c])
```

```python
# Iterable object = only configuration
# Iterator object = iteration state

# Instead of making one object track iteration, we separate responsibilities:

# ✔ Iterable → produces iterator
# ✔ Iterator → produces values

class CounterIterator:

    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):

        if self.current < self.limit:
            self.current += 1
            return self.current

        raise StopIteration

class Counter:

    def __init__(self, limit):
        self.limit = limit

    def __iter__(self):
        return CounterIterator(self.limit)
    
c = Counter(3)

print([i for i in c])
print([i for i in c])
```
