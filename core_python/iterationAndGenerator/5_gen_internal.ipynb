A generator has four possible states.
| State         | Meaning                           |
| ------------- | --------------------------------- |
| GEN_CREATED   | Generator created but not started |
| GEN_RUNNING   | Currently executing               |
| GEN_SUSPENDED | Paused at `yield`                 |
| GEN_CLOSED    | Finished execution                |


```python
def gen():
    x = 10
    y = 20
    yield x + y

g = gen()

print(g.gi_frame.f_locals)
next(g)
print(g.gi_frame.f_locals)

# The frame stores:
# local variables
# instruction pointer
# execution stack
```

Generator Control Methods
| Method             | Purpose                  |
| ------------------ | ------------------------ |
| `next()`           | Resume execution         |
| `send(value)`      | Send data into generator |
| `throw(exception)` | Inject exception         |
| `close()`          | Stop generator           |


```python
# send(value) resumes the generator and injects a value into the paused yield expression.

# Important Rule
# First call must be next().

def gen():
    x = yield
    print(x)
    yield

g = gen()
next(g)
g.send(12)
```

```python
# throw() — Inject Exception Into Generator
# throw() allows caller to raise an exception inside the generator.
# g.throw(ValueError) is equivalent to raise ValueError inside the generator.

def gen():
    try:
        yield 1
    except ValueError:
        print("handled")

g = gen()

next(g)
g.throw(ValueError)
```

```python
def worker():
    while True:
        try:
            value = yield
            print("Processing", value)
        except ValueError:
            print("Invalid value")

g = worker()

next(g)

g.send(10)
g.throw(ValueError)
g.send(20)
```

```python
def gen():
    try:
        yield 1
    finally:
        print("cleanup")

g = gen()

next(g)
g.close()
# close() terminates the generator.
# Generators run finally blocks on close.
```
