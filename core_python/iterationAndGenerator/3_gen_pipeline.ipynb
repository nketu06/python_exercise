What is a Generator Pipeline?
A pipeline = Chain of generators where output of one feeds next.

```python
nums = range(10)

squared = (x*x for x in nums)
evens = (x for x in squared if x % 2 == 0)

print(list(evens))

# ✅ No large lists
# ✅ Lazy evaluation
# ✅ Values generated on demand
# saves form memory explosion for larger numbers like 1000000000
# range → square → filter → output
```

```python
def square(nums):
    for x in nums:
        yield x*x

def filter_even(nums):
    for x in nums:
        if x % 2 == 0:
            yield x
# Generator functions
```

```python
def square(nums):
    print("Squaring...")
    for x in nums:
        yield x*x

nums = square(range(3))

print(next(nums))
print(next(nums))
```

```python
def infinite_numbers():
    ans = 0
    while True:
        ans+=1
        yield ans

pipeline = square(infinite_numbers())

# print(list(pipeline))
# dangerous
```
