What yield from REALLY Does

It automatically handles:

✔ Iteration
✔ Passing values via send()
✔ Propagating exceptions
✔ Returning values from sub-generator

Equivalent to writing:
def outer():
    for value in inner():
        yield value

```python
def inner():
    yield 1
    yield 2

def outer():
    yield from inner()

print(list(outer()))
```

```python
def inner():
    x = yield
    yield x * 2

def outer():
    yield from inner()

g = outer()
next(g)
print(g.send(10))
```

```python
# Clean Flattening Pattern
def flatten(list_of_lists):
    for sublist in list_of_lists:
        yield from sublist
print(list(flatten([[1,2], [3,4]])))
```

```python
def inner():
    yield 1
    return 5

def outer():
    x = yield from inner()
    yield x

print(list(outer()))
```
