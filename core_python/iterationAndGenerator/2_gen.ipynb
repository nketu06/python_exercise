```python
# A generator is a function that pauses execution and resumes later.

# Memory efficiency
# ✔ Lazy evaluation

# Generator Core Takeaways (MEMORIZE)
# ✔ yield = pause + resume
# ✔ Generator object ≠ function execution
# ✔ return inside generator → StopIteration
# ✔ Lazy evaluation vs eager
# ✔ next() sends None. next(g) = send(None)
# ✔ Generator = state machine
```

```python
def gen():
    print("Start")
    yield 1
    print("End")

g = gen()
print(g)
```

```python
next(g) # pause at yield
```

| Feature          | return | yield |
| ---------------- | ------ | ----- |
| Ends function?   | ✅ Yes  | ❌ No  |
| Saves state?     | ❌ No   | ✅ Yes |
| Multiple values? | ❌ No   | ✅ Yes |
| Lazy evaluation? | ❌ No   | ✅ Yes |


```python
def counter():
    x = 0
    while True:
        x += 1
        yield x

c=counter()

# Each next()
# ✔ Resumes
# ✔ Continues
# ✔ Pauses again

# -----generator expression--------
# (x for x in iterable) not in sq bracket

# Generator (Lazy Evaluation)
# HUGE difference in large systems.
nums = (x*x for x in range(1000000))
```

```python
# Generator remembers:
# ✔ Local variables
# ✔ Execution position
# ✔ Stack frame

print(c.__next__()) # it will increment as you run

print(next(nums))
```

```python
# sending value in generator
def gen():
    print("Waiting...")
    x = yield
    print("Received:", x)

g = gen()

next(g)
g.send(100)
```

A generator must be primed before sending values. Calling next() advances execution to the first yield, allowing send() to inject data into the paused generator.
