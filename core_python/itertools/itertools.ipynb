itertools = Lego blocks for building iteration pipelines.

All functions return iterators, not lists.

So they are:

✔ Lazy
✔ Memory-efficient
✔ Fast (implemented in C)

```python
from itertools import count

for i in count(10):
    if i > 15:
        break
    print(i)

# ✔ Infinite counter
# ✔ Useful for indexing streams
```

```python
from itertools import cycle

c = cycle([1, 2, 3])
for _ in range(7):
    print(next(c))
```

```python
from itertools import repeat

for x in repeat("A", 3):
    print(x)
```

```python
from itertools import chain

for x in chain([1,2], [3,4]):
    print(x)
```

🐍 Python itertools — Complete Practical Notes

itertools provides fast, memory-efficient iterator building blocks implemented in C.

All functions return iterators (lazy evaluation) unless explicitly converted using list().

📌 Categories of itertools

Infinite Iterators

Finite Iterators

Combinatoric Iterators

Useful Utility Patterns

1️⃣ Infinite Iterators

⚠️ Use carefully — can run forever.

🔹 count(start=0, step=1)

Returns evenly spaced values starting from start.

from itertools import count

for i in count(10):
    if i > 15:
        break
    print(i)

Output:

10 11 12 13 14 15
Use Cases:

Generating IDs

Infinite streams

Pairing with zip()

🔹 cycle(iterable)

Repeats elements indefinitely.

from itertools import cycle

c = cycle([1, 2, 3])
for _ in range(7):
    print(next(c))

Output:

1 2 3 1 2 3 1
Use Cases:

Round-robin scheduling

Repeating patterns

🔹 repeat(object, times=None)

Repeats a value.

from itertools import repeat

list(repeat("A", 3))

Output:

['A', 'A', 'A']
Use Cases:

Default values

Filling structures

2️⃣ Finite Iterators (Most Practical)
🔹 chain(*iterables)

Combines multiple iterables lazily.

from itertools import chain

list(chain([1,2], [3,4]))

Output:

[1,2,3,4]
Why Better Than +?

No intermediate list creation

Memory efficient

🔹 islice(iterable, start, stop, step)

Lazy slicing.

from itertools import islice

nums = range(100)
list(islice(nums, 5))

Output:

[0,1,2,3,4]
Use Case:

Slicing large generators safely

🔹 filterfalse(predicate, iterable)

Opposite of filter().

from itertools import filterfalse

nums = [1,2,3,4]
list(filterfalse(lambda x: x%2==0, nums))

Output:

[1,3]
🔹 takewhile(predicate, iterable)

Takes values while condition is true.

from itertools import takewhile

list(takewhile(lambda x: x<5, range(10)))

Output:

[0,1,2,3,4]
🔹 dropwhile(predicate, iterable)

Drops values while condition is true.

from itertools import dropwhile

list(dropwhile(lambda x: x<5, range(10)))

Output:

[5,6,7,8,9]
🔹 accumulate(iterable, func=operator.add)

Cumulative operation.

from itertools import accumulate

list(accumulate([1,2,3,4]))

Output:

[1,3,6,10]
Use Cases:

Running totals

Prefix sums

🔹 groupby(iterable, key=None)

Groups consecutive items.

from itertools import groupby

data = [1,1,2,2,2,3]
for key, group in groupby(data):
    print(key, list(group))

Output:

1 [1,1]
2 [2,2,2]
3 [3]

⚠️ Requires sorted input for proper grouping.

3️⃣ Combinatoric Iterators (Interview Gold ⭐⭐⭐⭐⭐)
🔹 product(*iterables, repeat=1)

Cartesian product.

from itertools import product

list(product([1,2], ['A','B']))

Output:

(1,'A'), (1,'B'), (2,'A'), (2,'B')

Equivalent to nested loops.

🔹 permutations(iterable, r=None)

Order matters.

from itertools import permutations

list(permutations([1,2,3], 2))

Output:

(1,2), (1,3), (2,1), (2,3), (3,1), (3,2)
🔹 combinations(iterable, r)

Order does NOT matter.

from itertools import combinations

list(combinations([1,2,3], 2))

Output:

(1,2), (1,3), (2,3)
🔹 combinations_with_replacement(iterable, r)

Allows repetition.

from itertools import combinations_with_replacement

list(combinations_with_replacement([1,2], 2))

Output:

(1,1), (1,2), (2,2)
4️⃣ Performance & Memory Notes ⭐⭐⭐⭐⭐
Key Principles:

All itertools functions are lazy

Memory usage = O(1) (unless converted to list)

Implemented in C → very fast

Best for large data processing

🔥 Interview-Level Insights

chain() is better than + for large iterables

product() replaces nested loops cleanly

groupby() requires sorted input

count() must be sliced (dangerous infinite iterator)

islice() is essential for controlling infinite streams

🧠 Mental Model

itertools = Functional Lego blocks for building iteration pipelines.

Instead of writing:

for a in A:
    for b in B:

You write:

for a, b in product(A, B):

Cleaner, faster, more Pythonic.
