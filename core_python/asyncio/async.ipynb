async def creates coroutine

calling coroutine does not execute

await suspends coroutine

event loop schedules tasks

async best for I/O bound problems

```python
import asyncio

async def f():
    print("done")
    await asyncio.sleep(3)
    print("done")

# asyncio.run(f())
```

async function
      ↓
returns coroutine object
      ↓
event loop schedules it
      ↓
await pauses execution
      ↓
resume later
