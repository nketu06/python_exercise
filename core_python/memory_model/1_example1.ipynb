Everything in Python is an Object  
numbers → objects  
strings → objects  
lists → objects  
functions → objects  

```python
x = 10
y = 10

print(id(x))
print(id(y))
# variables do not store values. They store references to objects.
# Sometimes same, sometimes not (due to optimization).
```

Immutable
Cannot change after creation.  

int  
float  
str  
tuple  
frozenset  

```python
x = 10
print(id(x))
x = x + 1
print(id(x))
# Actually creates new object.
```

Mutable  
list  
dict  
set  
bytearray  

```python
a = [1,2]
print(id(a))
a.append(3)
print(id(a))
```

Shallow Copy and Deep Copy

```python
a =[]

b=a
# same list

import copy
b = copy.copy(a)
# New list, but nested objects shared.

b = copy.deepcopy(a)
# Completely independent copy.
```

```python
# Mutable Default Argument Trap
def add_item(x, lst=[]):
    lst.append(x)
    return lst
print(add_item(1))
print(add_item(2))

# Why?
# Default argument evaluated once at function definition.

# Correct Pattern
# def add_item(x, lst=None):
#     if lst is None:
#         lst = []
#     lst.append(x)
#     return lst
```

```python
# Python caches small integers.
# Typically: -5 to 256
# Because Python reuses same object.
a = 10
b = 10

print(a is b)

a = 2984587289
b = 698428592

print(a is b)

```

1️⃣ Variables store references, not values
