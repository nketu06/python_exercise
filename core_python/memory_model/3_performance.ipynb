Why Python Code Can Be Slow  

Python is interpreted and dynamically typed.

```python
for i in range(1_000_000):
    x = i * i

# Each loop iteration requires:
    # 1. bytecode execution
    # 2. dynamic type checks
    # 3. object creation
# So Python loops are slow compared to C loops.
```

2️⃣ Avoid Python Loops When Possible

Bad:

result = []  
for x in nums:  
    result.append(x*x)  

Better:  

result = [x*x for x in nums]  

Why faster?  

list comprehension runs in C internally

```python
# Use Built-in Functions

# Built-ins are implemented in C, so they are faster.

# Example:

# Bad:
total = 0
for x in nums:
    total += x

# Better:
total = sum(nums)
```

```python
# Use Generators for Large Data
nums = (x*x for x in range(10_000_000))
```

```python
# Local Variables Are Faster

# Accessing local variables is faster than globals.

# Example:

# Slow:
import math
for i in range(1000000):
    math.sqrt(i)

# Better:
sqrt = math.sqrt

for i in range(1000000):
    sqrt(i)

# Because Python avoids repeated attribute lookup.
```

```python
# 7️⃣ __slots__ Saves Memory

# Normal object:
# object + dictionary

# With slots:
# object only
# Example:

class User:
    __slots__ = ['name', 'age']

# Useful when creating millions of objects.
```

```python
# Profiling Code
# Never optimize blindly.
# Use profiling tools.
# timeit

# Example:

import timeit
timeit.timeit("sum(range(1000))", number=1000) 

# cProfile
# Example:

import cProfile
cProfile.run("my_function()")

# Shows where time is spent.
```

Golden Rule  
Algorithm > Data Structure > Python tricks.  
A better algorithm beats micro-optimizations.
